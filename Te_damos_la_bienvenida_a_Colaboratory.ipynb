```
#Parte 1#
#Ejercicio 1#

def operaciones_basicas(num1, num2):
    suma = num1 + num2
    resta = num1 - num2
    multiplicacion = num1 * num2
    return suma, resta, multiplicacion

# Ejemplo de uso:
numero1 = int(input("Ingrese el primer número: "))
numero2 = int(input("Ingrese el segundo número: "))

resultado_suma, resultado_resta, resultado_multiplicacion = operaciones_basicas(numero1, numero2)

print(f"La suma de {numero1} y {numero2} es: {resultado_suma}")
print(f"La resta de {numero1} y {numero2} es: {resultado_resta}")
print(f"La multiplicación de {numero1} y {numero2} es: {resultado_multiplicacion}")
```

```
#Parte 1#
#Ejercicio 2#

def analizar_lista_numeros(lista_numeros):
    if not lista_numeros:
        return None, None, None # Retorna None si la lista está vacía

    numero_mayor = max(lista_numeros)
    numero_menor = min(lista_numeros)
    promedio = sum(lista_numeros) / len(lista_numeros)
    return numero_mayor, numero_menor, promedio

# Ejemplo de uso:
mi_lista = []
print("Ingresa los números para la lista (escribe 'fin' para terminar):")
while True:
    entrada = input("Ingresa un número: ")
    if entrada.lower() == 'fin':
        break
    try:
        numero = int(entrada)
        mi_lista.append(numero)
    except ValueError:
        print("Entrada inválida. Por favor, ingresa un número entero o 'fin'.")

mayor, menor, prom = analizar_lista_numeros(mi_lista)

if mayor is not None:
    print(f"La lista es: {mi_lista}")
    print(f"El número mayor es: {mayor}")
    print(f"El número menor es: {menor}")
    print(f"El promedio es: {prom}")
else:
    print("La lista está vacía.")
```
